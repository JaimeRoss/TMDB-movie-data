```python
#imports
import pandas as pd
import numpy as np
import matplotlib as pyplot
import seaborn as sns

# suppress scientific notation by setting float_format
pd.options.display.float_format = '{:.0f}'.format

# Setting the warnings to be ignored 
import warnings 
warnings.filterwarnings('ignore')
```

```python
# loading in data for year 2000
df2000 = pd.read_csv("Data/final_tmdb_data_2000.csv.gz")
df2000.head()
```

```python
df2000.shape
```

```python
# loading in data for year 2001
df2001 = pd.read_csv("Data/final_tmdb_data_2001.csv.gz")
df2001.head()
```

```python
df2001.shape
```

```python
# concatenating dataframes
dfcombined = pd.concat([df2000,df2001], axis = 0)
dfcombined.head()
```

```python
dfcombined.shape
```

How many movies had at least some valid financial information (values > 0 for budget OR revenue)?

```python
dfvalidfininfo = dfcombined.loc[(dfcombined["revenue"] > 0.0) | (dfcombined["budget"] > 0.0)]
dfvalidfininfo.head()
```

```python
dfvalidfininfo["imdb_id"].value_counts().sum()
```

There were 756 entries that had some valid financial information.

How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
dfvalidfininfo["certification"].value_counts()
```

```python
ax = sns.countplot(data = dfvalidfininfo, x="certification")
ax.set_title("Countplot for values of certification");
```

```python
dfvalidfininfo["certification"].isna().sum().sum()
```

There are 319 missing values for the certification column.

What is the average revenue per certification category?

```python
tempdf = dfvalidfininfo.groupby(["certification"]).mean()
tempdf[["revenue"]]
```

```python
ax = sns.barplot(data = dfvalidfininfo, x="certification", y = "revenue")
ax.set_title("Countplot for average revenues grouped by certification");
```

What is the average budget per certification category?

```python
tempdf[["budget"]]
```

```python
ax = sns.barplot(data = dfvalidfininfo, x="certification", y = "budget")
ax.set_title("Countplot for average budget grouped by certification");
```

Saving final dataframe as a CSV file

```python
dfvalidfininfo.to_csv('Data/tmdb_results_combined.csv.gz', index=False, compression='gzip')
```
